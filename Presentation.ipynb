Må lage kategoriske variable

````python
categories = ['job', 'marital', 'education', 'contact', 'month', 'poutcome']
for var in categories:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(df[var], prefix=var)
    df = df.join(cat_list)
````

Tilsvarende for boolske variable

````python
bools = ['default', 'housing', 'loan', 'y']

def boolean(df, name):
    df[name] = (df[name] == 'yes').astype(int)
    return df

for var in bools:
    df = boolean(df, var)
````

````python
print(df.groupby("y").mean())
````

Hvilken tolkning har månedene her?

````python
a = 0
months = ['jan','feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
for month in months:
    i = 'month_' + month
    a += df.groupby('y')[i].mean()
print(a)
````

````python
df = df.drop(['job_unemployed','marital_single','education_primary', \
'contact_cellular', 'month_jan', 'poutcome_failure'],1)

df['intercept'] = 1
````

Vil også fjerne originale kategoriske variable:

````python
variables = (df.columns.values.tolist())
keep = [i for i in variables if i not in categories]
data = df[keep]
````

Vi kan også se nærmere på sammenhengen mellom suksessrate og måneder. 

Jeg fant ingen intuitiv måte å gjøre dette i _Pandas_, men ved litt manipulasjon og bruk av dictionaries lar det seg gjøre å hente ut antall oppringninger og suksesser per måned:

````python
df = pd.read_csv('bank-full.csv', sep = ';', header=0)

df_lim = df[['day', 'month', 'y', 'duration']]

dict = {}

for i in range(len(df_lim)):
    m = df_lim.iloc[i].month
    if m not in dict.keys():
        dict[m] = {}
    d = df_lim.iloc[i].day
    if d not in dict[m].keys():
        dict[m][d] = [0,0,0]
    s = df_lim.iloc[i].y
    if s == 'yes':
        dict[m][d][1] += 1
    dict[m][d][0] += 1

    M = {}
for m in dict.keys():
    n = 0; s = 0;
    for d in dict[m].keys():
        n += dict[m][d][0]
        s += dict[m][d][1]
    M[m] = [s/n]
````

````python
months = {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,\
          'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12}

R = [[],[]]
i = 0
for keys in M:
    R[0].append((M[keys][0]))
    R[1].append(months[keys])
    i += 1

R = np.asarray(R)
R.sort(axis=0)
````

````python
plt.figure()
my_xticks = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
plt.xticks(range(1,13), my_xticks)
plt.bar(R[1,:], R[0,:])
plt.legend()
plt.xlabel('Month')
plt.ylabel('Rate of success')
plt.title('Success rate by month')
plt.show()
````

## Oppgave 2

_Lag en ny variabel som teller antall av oppkall per dag (en rad er ett oppkall), igjen antar vi ingen årseffekt._
* _Plot denne nye variablen mot suksessraten for dagen._
* _Hva mener du dette plottet viser, dra gjerne inn spørsmål 1._

Vi kan bygge videre på dictionarien som ble opprettet tidligere, og telle over alle dager i alle måneder, og lagre både N og S. 

````python
N = []
S = []

for m in dict.keys():
    for d in dict[m].keys():
        n = dict[m][d][0]
        s = dict[m][d][1]
        N.append(n)
        S.append(s)
````

````python
X = np.array([N,S])
plt.figure()
plt.scatter(X[0], X[1]/X[0], marker='x', s=2, color='black')
plt.xlabel('Number of calls')
plt.ylabel('Rate of success')
plt.title("Success rate by number of calls per day")
plt.show()
````
